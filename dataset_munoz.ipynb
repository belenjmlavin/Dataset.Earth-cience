**Dataset Climate Change**

```python
# Importar librerías
import pandas as pd

# Cargar el dataset
climate_df = pd.read_csv('/content/Datasets Climate Change/climate_change_dataset.csv')

# Mostrar las primeras filas del dataset
print("Climate Change Dataset:")
print(climate_df.head())

# Descripción de las variables del Climate Change Dataset
print("\nDescripción de variables - Climate Change Dataset:")
print(climate_df.describe(include='all'))  # Incluye todas las columnas, incluso no numéricas

# Describir las variables interesantes
print("\nVariables interesantes:")
print("1. Year: Año de la medición, importante para análisis de tendencias a lo largo del tiempo.")
print("2. Month: Mes de la medición, útil para análisis estacionales.")
print("3. Avg_Temp (°C): Temperatura promedio, clave para entender el calentamiento global.")
print("4. Max_Temp (°C): Temperatura máxima, puede indicar eventos climáticos extremos.")
print("5. Min_Temp (°C): Temperatura mínima, útil para análisis de variabilidad climática.")
print("6. Precipitation (mm): Precipitación, esencial para estudios de sequías e inundaciones.")
print("7. Humidity (%): Humedad, importante para entender el clima y el confort humano.")
print("8. Wind_Speed (m/s): Velocidad del viento, relevante para estudios de energía eólica y clima.")
print("9. Solar_Irradiance (W/m²): Irradiancia solar, importante para estudios de energía solar.")
print("10. Cloud_Cover (%): Cobertura de nubes, relevante para estudios climáticos.")
print("11. CO2_Concentration (ppm): Concentración de CO2, un indicador crítico del cambio climático.")
print("12. Latitude: Latitud geográfica, útil para análisis regionales.")
print("13. Longitude: Longitud geográfica, útil para análisis regionales.")
print("14. Altitude (m): Altitud, relevante para estudios de clima en diferentes elevaciones.")
print("15. Proximity_to_Water (km): Proximidad al agua, importante para estudios de clima costero.")
print("16. Urbanization_Index: Índice de urbanización, relevante para estudios de impacto urbano.")
print("17. Vegetation_Index: Índice de vegetación, útil para estudios de ecosistemas.")
print("18. ENSO_Index: Índice de Oscilación del Sur de El Niño, importante para estudios climáticos.")
print("19. Particulate_Matter (µg/m³): Materia particulada, relevante para estudios de calidad del aire.")
print("20. Sea_Surface_Temp (°C): Temperatura de la superficie del mar, importante para estudios oceánicos.")
```
